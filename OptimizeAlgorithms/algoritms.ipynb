## TSP (Traveling Salesman Problem Implementation)

```python
graph = [[0, 10, 15, 20], [10, 0, 35, 25], 
        [15, 35, 0, 30], [20, 25, 30, 0]]
visualize_graph2() 
```

```python
tsp_solver = TSPSolver(graph)
print(f'Cost value: {tsp_solver.solve()}')
print("Complexity of TSPSolver is O(n!)")
print(f"Path: {tsp_solver.print_path()}")
```

```python
greedy_solver = GreedySolver(graph)
print(f'Cost value: {greedy_solver.solve()}')
print("Complexity of GreedySolver is O(n^2)")
print(f"Path: {greedy_solver.print_path()}")
```

```python
dynamic_solver = DynamicSolver(graph)
print(f'Cost value: {dynamic_solver.solve()}')
print("Complexity of DynamicSolver is O(n^2 * 2^n)")
```

```python
print("----------------------------------")
bubble_sort = BubbleSort()
print(bubble_sort.sort([5, 4, 3, 2, 1]))
print("Complexity of BubbleSort is O(n^2)")
print("----------------------------------")

print("----------------------------------")    
selection_sort = SelectionSort()
print(selection_sort.sort([5, 4, 3, 2, 1]))
print("Complexity of SelectionSort is O(n^2)")
print("----------------------------------")

print("----------------------------------")    
insertion_sort = InsertionSort()
print(insertion_sort.sort([5, 4, 3, 2, 1]))
print("Complexity of InsertionSort is O(n^2)")
print("----------------------------------")
    
print("----------------------------------")
merge_sort = MergeSort()
print(merge_sort.sort([5, 4, 3, 2, 1]))
print("Complexity of MergeSort is O(n log n)")
print("----------------------------------")
    
print("----------------------------------")
quick_sort = QuickSort()
print(quick_sort.sort([5, 4, 3, 2, 1], 0, 4))
print("Complexity of QuickSort is O(n^2)")
print("----------------------------------")
    
print("----------------------------------")
heap_sort = HeapSort()
print(heap_sort.sort([5, 4, 3, 2, 1]))
print("Complexity of HeapSort is O(n log n)")
print("----------------------------------")

print("----------------------------------")
counting_sort = CountingSort()
print(counting_sort.sort([5, 4, 3, 2, 1]))
print("Complexity of CountingSort is O(n + k)")
print("----------------------------------")
    
print("----------------------------------")
radix_sort = RadixSort()
print(radix_sort.sort([5, 4, 3, 2, 1]))
print("Complexity of RadixSort is O(n * k)")
print("----------------------------------")
    
print("----------------------------------")
bucket_sort = BucketSort()
print(bucket_sort.sort([5, 4, 3, 2, 1]))
print("Complexity of BucketSort is O(n)")
print("----------------------------------")
    
print("----------------------------------")
shell_sort = ShellSort()
print(shell_sort.sort([5, 4, 3, 2, 1]))
print("Complexity of ShellSort is O(n^2)")
print("----------------------------------")
    
print("----------------------------------")
cocktail_sort = CocktailSort()
print(cocktail_sort.sort([5, 4, 3, 2, 1]))
print("Complexity of CocktailSort is O(n^2)")
print("----------------------------------")
```

```python
binary_search = BinarySearch()
print(binary_search.search([1, 2, 3, 4, 5], 3))
print("Complexity of BinarySearch is O(log n)")
    
linear_search = LinearSearch()
print(linear_search.search([1, 2, 3, 4, 5], 3))
print("Complexity of LinearSearch is O(n)")
 
jump_search = JumpSearch()
print(jump_search.search([1, 2, 3, 4, 5], 3))
print("Complexity of JumpSearch is O(√n)")
   
interpolation_search = InterpolationSearch()
print(interpolation_search.search([1, 2, 3, 4, 5], 3))
print("Complexity of InterpolationSearch is O(log log n)")
    
exponentials_search = ExponentialsSearch()
print(exponentials_search.search([1, 2, 3, 4, 5], 3))
print("Complexity of ExponentialsSearch is O(log n)")
```
