```python
import scipy.optimize as opt
import numpy as np
import matplotlib.pyplot as plt

def func_x(x):
    return - x**5 - 2*x + 1

x = np.arange(-10, 10, 0.1)
plt.plot(x, func_x(x))
```

# Single-variable function

```python
%%time
Newton_Raphson = opt.newton(func_x, x0=0.1)
print("Newton_Raphson: ", Newton_Raphson)
```

```python
%%time
Brent = opt.brentq(func_x, a=-10, b=10)
print("Brent: ", Brent)
```

```python
%%time
Bisect = opt.bisect(func_x, a=-10, b=10)
print("Bisect: ", Bisect)
```

# Multi-variable function

```python
def func_xy(x, y):
    return x**3 + (x)**2 + (y) - 4

x = np.arange(-10, 10, 0.1)
y = np.arange(-10, 10, 0.1)

X, Y = np.meshgrid(x, y)
Z = func_xy(X, Y)  

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)
ax.contour(X, Y, Z, levels=[0])
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_aspect('equal')
ax.grid(True)
```

```python
%%time
Newton_Raphson = opt.newton(func_xy, x0=0.1, args=(0.1,))
print("Newton_Raphson: ", Newton_Raphson)
```

```python
%%time
Brent = opt.brentq(func_xy, a=-10, b=10, args=(0.1,))
print("Brent: ", Brent)

```

```python
%%time
Bisect = opt.bisect(func_xy, a=-10, b=10, args=(0.1,))
print("Bisect: ", Bisect)
```
